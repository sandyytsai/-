```python
import pandas as pd
import datetime

# 1. 電價對照表 (Tariff Price)
TARIFF_MAP = {
    "P1": 0.18197, "P2": 0.13539, "P3": 0.09778,
    "P4": 0.07912, "P5": 0.06138, "P6": 0.08929
}

# 2. 每年固定國定假日 (Month, Day)
FIXED_HOLIDAYS = [
    (1, 1), (1, 6), (4, 18), (5, 1), (8, 15),
    (10, 12), (11, 1), (12, 6), (12, 8), (12, 25)
]

# 3. 全年度時段矩陣配置 (工作日規則)
MATRIX = {
    "P6_HOURS": list(range(0, 8)),  # 00:00 - 07:59 一律 P6
    "MONTHS": {
        # 高價月: P1 尖峰
        1:  {"P2": [8, 14, 15, 16, 17, 22, 23], "P1": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        2:  {"P2": [8, 14, 15, 16, 17, 22, 23], "P1": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        7:  {"P2": [8, 14, 15, 16, 17, 22, 23], "P1": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        12: {"P2": [8, 14, 15, 16, 17, 22, 23], "P1": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        # 中價月: P2 尖峰
        3:  {"P3": [8, 14, 15, 16, 17, 22, 23], "P2": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        11: {"P3": [8, 14, 15, 16, 17, 22, 23], "P2": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        # 中低價月: P3 尖峰
        6:  {"P4": [8, 14, 15, 16, 17, 22, 23], "P3": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        8:  {"P4": [8, 14, 15, 16, 17, 22, 23], "P3": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        9:  {"P4": [8, 14, 15, 16, 17, 22, 23], "P3": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        # 低價月: P4 尖峰
        4:  {"P5": [8, 14, 15, 16, 17, 22, 23], "P4": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        5:  {"P5": [8, 14, 15, 16, 17, 22, 23], "P4": [9, 10, 11, 12, 13, 18, 19, 20, 21]},
        10: {"P5": [8, 14, 15, 16, 17, 22, 23], "P4": [9, 10, 11, 12, 13, 18, 19, 20, 21]} # Modified P4 for October
    }
}

def get_period(date_obj, hour_val):
    # 1. Check if it is a Holiday (Weekend or Fixed Holiday)
    is_weekend = date_obj.weekday() >= 5 # 5=Sat, 6=Sun
    is_fixed_holiday = (date_obj.month, date_obj.day) in FIXED_HOLIDAYS

    if is_weekend or is_fixed_holiday:
        return "P6"

    # 2. Workday Logic
    # 00:00 - 07:59 are always P6
    if hour_val in MATRIX["P6_HOURS"]:
        return "P6"

    # Get the month's tariff configuration
    month_config = MATRIX["MONTHS"].get(date_obj.month)

    if month_config:
        # Iterate through the periods defined for the current month
        # and return the first matching period for the given hour.
        # The order of checking depends on how the MATRIX is structured.
        # Assuming MATRIX lists higher-priced periods first for specific hours.
        for period, hours_list in month_config.items():
            if hour_val in hours_list:
                return period

    # If the hour is not in P6_HOURS and not in any specific month's periods (8-23),
    # it defaults to P6 as per common tariff structures for undefined hours.
    # Based on the current MATRIX, all hours from 8 to 23 should be covered for workdays.
    return "P6"
```

```python
import datetime

# Example date and hour to test
example_date = datetime.date(2024, 10, 15) # Monday, October 15, 2024
example_hour = 22 # 22:00 PM

# Get the tariff period for the example date and hour
tariff_period = get_period(example_date, example_hour);

# Get the price from TARIFF_MAP
tariff_price = TARIFF_MAP.get(tariff_period);

print(f"For {example_date.strftime('%Y-%m-%d')} at {example_hour:02d}:00:")
print(f"Tariff Period: {tariff_period}")
print(f"Corresponding Price: {tariff_price:.5f} (EUR/kWh)")
```
